A function $f$ is a rule that assigns to each element $x$ in a set $A$ exactly one element, called $f(x)$, in set $B$.

* independent variable
* dependent variable

```python
# absolute value
%matplotlib inline
def f(x):
    if x >=0.0:
        return x
    else:
        return -x
  
x = linspace(-10,10,1000)

y=[]
for x_i in x:
    y.append(f(x_i))

plot(x,y,label='y=|x|')
legend()
```

```python
# Symmetry
## even function
## odd function
x = linspace(-pi,pi,1000)
y1 = sin(x)
y2 = cos(x)


ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'])

plot(x,y1)
plot(x,y2)
```

```python
def plot_xy(x,y):
    
    ax = gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    
    plot(x,y)
```

```python
x = linspace(-pi,pi,1000)
y1 = sin(x)
y2 = cos(x)
plot_xy(x,y1)
plot_xy(x,y2)
```

Increasing and Decresing Functions

* increasing: $f(x_1) < f(x_2)$ whenever $x_1 < x_2$
* decreasing: $f(x_1) > f(x_2)$ whenever $x_1 < x_2$
